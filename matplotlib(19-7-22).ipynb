```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.array([1, 2, 6, 7])
y = np.array([3, 8, 1, 10])
plt.plot(x, y)
plt.show()
```

```python
# Default x points
y = np.array([3, 4, 1, 6, 2])
plt.plot(y)
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o')
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, 'o:r') # 'o' for marker, ':' for dotted line, 'r' for red color
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o', ms = 20)
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 's', mec = 'r')
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'D', mfc = 'r')
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
x = np.array([1, 2, 6, 7])
y = np.array([3, 8, 1, 10])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('SALES')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title('INCOME')

plt.suptitle('SHOP')
plt.show()
```
