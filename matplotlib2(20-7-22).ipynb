```python
import numpy as np
arr = np.array([10, 20, 30, 50, 90])
print(np.mean(arr))
print(np.var(arr))
print(np.std(arr))
```

```python
# Scatter plot
import numpy as np
import matplotlib.pyplot as plt
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
plt.scatter(x, y, color = 'hotpink')
plt.show()
```

```python
# Bar plot
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = 'orange')
plt.show()
```

```python
plt.barh(x, y, color = 'green')
plt.show()
```

```python
# Histogram
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
#Pie charts
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
mylabels = ['Apple', 'Banana', 'Cherry', 'Grapes']
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
mycolors = ['black', 'hotpink', 'r', 'b']
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
mycolors = ['black', 'hotpink', 'r', 'b']
plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend(title = 'FRUITS')
plt.show()
```
